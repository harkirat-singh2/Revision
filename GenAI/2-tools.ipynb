```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_groq import ChatGroq

model = ChatGroq(
    model ="llama-3.3-70b-versatile",
    api_key=os.getenv("GROQ_API_KEY")
)

response = model.invoke("2 points about AI")
print(response.content)
```

```python
#  tools

from langchain.tools import tool

@tool
def get_weather(location:str)->str:
  """Get weather at a locaion"""
  return f"It's a sunny {location}"

model_with_tools = model.bind_tools([get_weather])

response =  model_with_tools.invoke("What is the wather in Lucknow")

print(response)
```

```python
response.tool_calls
```

```python
# Step 1: Model generates tool calls
messages = [{"role": "user", "content": "What's the weather in Bangalore?"}]

ai_msg = model_with_tools.invoke(messages)
messages.append(ai_msg)

# Step 2: Execute tools
for tool_call in ai_msg.tool_calls:
    tool_name = tool_call["name"]
    tool_args = tool_call["args"]

    # call your function properly
    tool_result = get_weather.invoke(tool_args)

    # append tool response in correct format
    messages.append({
        "role": "tool",
        "content": str(tool_result),
        "tool_call_id": tool_call["id"]
    })

# Step 3: Final response
final_response = model_with_tools.invoke(messages)

print(final_response.content)
```
