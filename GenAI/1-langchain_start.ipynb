### Langchain version V1


```python
import langchain
print(langchain.__version__)
```

```python
import os
from dotenv import load_dotenv
load_dotenv()

groq_api_key = os.getenv("GROQ_API_KEY")
```

### GenerativeAI Application


```python
from langchain.chat_models import init_chat_model
model = init_chat_model(
    "llama-3.3-70b-versatile",
    model_provider="groq"
)
```

```python
# Alternative Way
from langchain_groq import ChatGroq

model = ChatGroq(
    model="llama-3.3-70b-versatile",
    api_key=os.getenv("GROQ_API_KEY")  # MUST NOT be None
)
```

```python
response = model.invoke("5 lines about AI")
print(response.content)
```

### Streaming And Batch


```python
for chunk in model.stream("5 lines about AI"):
  print(chunk.text)
```

```python
response = model.batch(
    [
        "Who is president of India",
        "Who is president of Australia",
        "Who is president of China"
    ],
    config={
        "max_concurrency": 5
    }
)

for r in response:
    print(r.content)
```
